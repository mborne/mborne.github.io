## Basic example

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-np.pi, np.pi, 100)

plt.plot(x, np.cos( x ), label='Cosinus')
plt.plot(x, np.sin( x ), label='Sinus')
plt.legend()
plt.show()
````
